```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    def __and__(self,other):
        s = []
        for i in self:
            if i in other:
                s.append(i)
        return MySet.str(s)

    def __or__(self, other):
        s = []
        for i in self:
            s.append(i)
        for i in other:
            s.append(i)
        return MySet.str(s)

    def __sub__(self,other):
        s = []
        for i in self:
            if i not in other:
                s.append(i)
        return MySet.str(s)
       
    def str(l):
        result = "MySet: {"
        for e in l:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print (s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print (t)

    u = s | t
    print(u)

    u = s & t
    print(u)

    u = s - t
    print(u)
```
